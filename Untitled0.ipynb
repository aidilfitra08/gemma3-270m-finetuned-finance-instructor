```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_id = "google/gemma-3-270m"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    # load_in_4bit=True  # for Colab memory efficiency
)
```

```python
from datasets import load_dataset

dataset = load_dataset("Josephgflowers/Finance-Instruct-500k")

print(dataset["train"][0])
```

```python
def format_example(example):
    sys = example.get("system", "").strip()
    usr = example.get("user", "").strip()
    asp = example.get("assistant", "").strip()

    prompt = ""
    if sys:
        prompt += f"[SYSTEM]\n{sys}\n\n"
    if usr:
        prompt += f"[USER]\n{usr}\n\n"
    if asp:
        prompt += f"[ASSISTANT]\n{asp}"
    return {"text": prompt}

dataset = dataset.map(format_example, remove_columns=["system", "user", "assistant"])
```

```python
def tokenize_fn(examples):
    return tokenizer(examples["text"], truncation=True, padding="max_length", max_length=512)

tokenized = dataset.map(tokenize_fn, batched=True)
```

```python
from peft import LoraConfig, get_peft_model

lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["q_proj", "v_proj"],  # typically for transformer architectures
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
)

model = get_peft_model(model, lora_config)
```

```python
# Take 50k from train for faster testing
small_train = dataset["train"].shuffle(seed=42).select(range(50_000))

# Take 5k for evaluation
small_eval = dataset["train"].shuffle(seed=123).select(range(5_000))
```

```python
from transformers import TrainingArguments
from trl import SFTTrainer

training_args = TrainingArguments(
    output_dir="/content/drive/MyDrive/gemma3-finance-agent",  # where checkpoints & final model are stored
    per_device_train_batch_size=4,        # number of samples per GPU forward pass
    gradient_accumulation_steps=4,        # simulate bigger batch size (4×4=16 effective batch size)
    warmup_steps=1000,                     # learning rate warmup for first 100 steps
    num_train_epochs=3,                   # number of full passes through dataset
    learning_rate=1e-4,                   # starting LR for AdamW optimizer
    logging_dir="./logs",                 # folder for TensorBoard logs

    # ✅ Checkpoint options
    save_strategy="steps",   # when to save checkpoints ("steps" or "epoch")
    save_steps=5000,          # save a checkpoint every 5000 steps
    save_total_limit=2,      # keep only 2 latest checkpoints (old ones deleted)
    # evaluation_strategy="steps",  # when to run eval
    # eval_steps=5000,          # run evaluation every 5000 steps
    logging_steps=100,       # log training loss every 100 steps

    fp16=True,               # use half-precision (faster, less memory on GPUs)
    optim="adamw_torch",     # optimizer type (AdamW from PyTorch)
)

trainer = SFTTrainer(
    model=model,
    # train_dataset=tokenized["train"],
    # eval_dataset=tokenized["train"].train_test_split(test_size=0.05)["test"],  # small dev split
    train_dataset=small_train,
    eval_dataset=small_eval,
    # dataset_text_field="text",
    # tokenizer=tokenizer,
    args=training_args,
    # packing=True,
)

trainer.train()
```

```python
# Save locally
model.save_pretrained("./gemma3-finance-agent")
tokenizer.save_pretrained("./gemma3-finance-agent")
```

```python
from transformers import pipeline

agent = pipeline(
    "text-generation",
    model="./gemma3-finance-agent",
    tokenizer=tokenizer,
    device=0
)

print(agent("Explain the effects of increasing interest rates on bond prices.", max_new_tokens=200)[0]["generated_text"])
```
